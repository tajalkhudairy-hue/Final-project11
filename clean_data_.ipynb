```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/AmesHousing.csv")

def clean_data(df):
    df = df.copy()
    target = "SalePrice"
    df["PID"] = df["PID"].astype(str)
    df["MS SubClass"] = df["MS SubClass"].astype(str)
    num_cols = df.select_dtypes(include=[np.number]).columns
    cat_cols = df.select_dtypes(include=["object"]).columns
    df[num_cols] = df[num_cols].fillna(df[num_cols].median())
    df[cat_cols] = df[cat_cols].fillna("None")
    df = df.drop_duplicates()
    p99 = df[target].quantile(0.99)
    df[target] = df[target].clip(upper=p99)
    print("missing in saleprice:", df[target].isnull().sum())
    print("Duplicates:", df.duplicated().sum())
    print("min saleprice:", df[target].min())
    return df
df_cleaned = clean_data(df)
# Feature Engineering
df_features = df_cleaned.copy()

df_features["price_per_sqft"] = df_features["SalePrice"] / (df_features["Gr Liv Area"] + 1)
df_features["total_bathrooms"] = df_features["Full Bath"] + 0.5 * df_features["Half Bath"]
df_features["area_x_lot"] = df_features["Gr Liv Area"] * df_features["Lot Area"]

df_features["SalePrice_log"] = np.log1p(df_features["SalePrice"])
df_features["house_age"] = df_features["Yr Sold"] - df_features["Year Built"]
print(df_features.shape)
print(df_features.head())

# Save final dataset
df_features.to_csv("final_dataset.csv", index=False)

```
